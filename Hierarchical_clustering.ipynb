Hierarchical clustering is an **unsupervised learning** method that groups data into a **tree-like structure** called a **dendrogram**. It does not need a fixed number of clusters at the start. Clusters are formed step by step based on **distance or similarity** between data points. We can cut the dendrogram at any level to get clusters.

---

### ğŸ”¹ Types of Hierarchical Clustering

1. **Agglomerative (Bottom-Up)** ğŸ”¼

   * Each data point starts as its own cluster
   * Closest clusters are merged step by step
   * Most commonly used

2. **Divisive (Top-Down)** ğŸ”½

   * All data starts in one cluster
   * Clusters are split recursively
   * Less common, more complex

---

### ğŸ”¹ Linkage Methods (How distance is calculated)

* **Single linkage** â€“ nearest point distance
* **Complete linkage** â€“ farthest point distance
* **Average linkage** â€“ average distance
* **Ward linkage** â€“ minimizes variance (most popular)

---

### ğŸ”¹ Key Points to Remember

* No need to choose **K** first âŒ
* Uses **dendrogram** for visualization ğŸŒ³
* Works well for **small to medium datasets**
* Slower and memory-heavy for large data âš ï¸

---

### ğŸ”¹ Where it is Used

* Customer segmentation ğŸ›’
* Document & text clustering ğŸ“„
* Biology (genes, species) ğŸ§¬



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

# generate a random dataset
X , y = make_blobs(n_samples=100,centers=3,  n_features=3, random_state=42)

# preform hierarchical cluster with AgglomerativeClustering algorithm
cluster = AgglomerativeClustering(n_clusters=3,  linkage='ward')
cluster.fit_predict(X)

# plot the dendrogram
plt.figure(figsize=(10,7))
plt.title("Dendrogram")
dend = shc.dendrogram(shc.linkage(X, method='ward'))
plt.show()
```
